```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Load Dataset
df = pd.read_csv("energy_consumption.csv")

# Show first few rows
df.head()
```

```python
# 3. Dataset Overview
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())

# Descriptive statistics
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["Day"], df["Energy_Consumption"], marker='o', color="blue")
plt.title("Energy Consumption Over Days")
plt.xlabel("Day")
plt.ylabel("Energy Consumption (kWh)")
plt.grid(True)
plt.show()


```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Country", y="Renewable_Percentage", data=df, palette="viridis")
plt.title("Renewable Energy Percentage by Country")
plt.xlabel("Country")
plt.ylabel("Renewable Percentage (%)")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Energy_Consumption", y="CO2_Emissions", hue="Country", data=df, s=100)
plt.title("CO2 Emissions vs Energy Consumption")
plt.xlabel("Energy Consumption (kWh)")
plt.ylabel("CO2 Emissions (tons)")
plt.grid(True)
plt.show()
```

```python
avg_energy = df.groupby("Country")["Energy_Consumption"].mean()

plt.figure(figsize=(8,5))
avg_energy.plot(kind="bar", color="orange")
plt.title("Average Energy Consumption per Country")
plt.xlabel("Country")
plt.ylabel("Average Energy Consumption (kWh)")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Year", y="Energy_Consumption", hue="Country", data=df, marker="o")
plt.title("Energy Consumption Trend Over Years")
plt.xlabel("Year")
plt.ylabel("Energy Consumption (kWh)")
plt.grid(True)
plt.show()
```
